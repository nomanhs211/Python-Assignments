```python
dic = [
        
        {
            "Q." : "1. Hobbies?\n", 
            "Opt" : ["a) Cricket", "b) football", "c) Hockey", "d) Tennis"],
            "Ans" : "a"
        },
        {
            "Q." : "2. Type of Mobile?\n", 
            "Opt" : ["a) Samsung", "b) Infinix", "c) Oppo", "d) Nokia"],
            "Ans" : "b"
        },
        {
            "Q." : "3. color of Orange?\n", 
            "Opt" : ["a) Blue", "b) Black", "c) Red", "d) Green"],
            "Ans" : "c"
        },
        {
            "Q." : "4. Color of Van?\n", 
            "Opt" : ["a) Black", "b) Blue", "c) White", "d) Red"],
            "Ans" : "a"
        },
        {
            "Q." : "5. Your Education?\n", 
            "Opt" : ["a)Matric ", "b) Intermediate", "c) Graduation", "d) Masters"],
            "Ans" : "d"
        },   

       ]


l  = 0
for quiz in dic:
    print(quiz["Q."])
    for O in quiz["Opt"]:
        print(O)
    answer = input("answer: ")
    if answer == dic[l]["Ans"]:
        print("\nCorrect answer\n")
    else:
        print("\nWrong answer\n")
    l = l+1

      
  
```
